```python
x = [1,2,4,6,10,20]
```

```python
def power(lst,num):
    ret = []
    for i in lst:
        ret.append(i ** num)  #powering with 2
    return ret
```

```python
y_2 = power(lst=x,num=2)
y_3 = power(num=3,lst=x)
```

```python
y_2
```

```python
y_3
```

```python
plt.plot(x,y_2);
plt.plot(x,y_3);
plt.show()
```

```python
plt.plot(x,y_2)
plt.plot(x,y_3)
plt.xlabel='x values'
plt.ylabel='square of x'
plt.grid('on')
```

```python
plt.plot(x,y_2,'r*-')  # r: red , o:marker , ':'' line style
plt.plot(x,y_3,'g*-')
plt.xlabel='x values'
plt.ylabel='square of x'
plt.grid('on')
```

# Plot styles
* Colors : https://matplotlib.org/2.0.2/api/colors_api.html
* Markers: https://matplotlib.org/3.2.2/api/markers_api.html
* Line Styles: https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html

```python
plt.plot(x,y_2,'ro--', label='square')  # r: red , o:marker , ':'' line style
plt.plot(x,y_3,'g*:', label='cube')
plt.xlabel='x values'
plt.ylabel='square of x'
plt.grid('on')
plt.legend()
plt.title('A sample graph')
plt.show()
```

# PSUTIL
fetches the system resource information using python

```python
import psutil as ps
import time 
```

```python
for i in range(10):
    p = psutil.cpu_percent(interval=None, percpu=True)
    avg_load = sum(p)/len(p)
    print(f'CPU Usage is {avg_load}%')
    time.sleep(1)
```
